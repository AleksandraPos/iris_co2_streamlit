```python
import plotly.express as px
import pandas as pd
df = px.data.iris()
df.head()
```

```python
my_df = pd.read_csv('data/geo_data.csv')
my_df
```

```python
fig = px.scatter(df, x='sepal_width', y='sepal_length')
fig.show()
```

```python
df.columns
```

```python
fig = px.scatter(df, x='sepal_width',
                 y='sepal_length',
                 color='species',
                 size='petal_length',
                 hover_data=['petal_width'])
fig.show()
```

```python
fig = px.scatter(df, x='sepal_width',
                 y='sepal_length',
                 color='species',
                 size='petal_length',
                 hover_data=['petal_width'],
                 marginal_x='box')
fig.show()
```

```python
fig = px.scatter(df, x='sepal_width',
                 y='sepal_length',
                 color='species',
                 size='petal_length',
                 hover_data=['petal_width'],
                 marginal_x='box',
                 marginal_y = 'box')
fig.show()
```

```python
fig = px.scatter(df, x='sepal_width',
                 y='sepal_length',
                 color='species',
                 size='petal_length',
                 hover_data=['petal_width'],
                 marginal_x='histogram',
                 marginal_y = 'histogram'
                 ,trendline='ols'
                 )

#effacer la grille 
fig.update_layout(
    xaxis=dict(showgrid=False),
    yaxis=dict(showgrid=False)
)

fig.show()
```

```python
#histogram

fig = px.histogram(df, x="sepal_width")
fig
```

```python
#Violin plot 


fig = px.violin(df, y="sepal_width", color="species_id")
fig.show()
```

```python
fig = px.violin(df
                , y="sepal_width"
                , color="species"
                , box=True
                , points='all')
fig.show()
```

```python
df.groupby('species').count().reset_index()
```

```python
df.groupby('species').sum().reset_index()
```

```python
#Pie chart 

fig = px.pie(df.groupby('species').count().reset_index(), 
             values='sepal_length', names='species', 
             title='Propotion Cat')
fig.show()
```

```python
fig = px.pie(df.groupby('species').sum().reset_index(), 
             values='sepal_length', names='species', 
             title='Total sepal length by species')
fig.show()
```

# 3D PLot

```python
fig = px.scatter_3d(df, x='sepal_length', y='sepal_width', 
                    z='petal_width', color='species')
fig.show()
```

```python
# Maps 


co2 = pd.read_csv('data/CO2_per_capita.csv', sep=";")

co2 = co2.dropna().sort_values('Year')

# TODO: Visualize your data on a World map
fig = px.scatter_geo(co2, locations="Country Code",
                    # color="continent", # which column to use to set the color of markers
                    hover_name="Country Name", # column added to hover information
                    size="CO2 Per Capita (metric tons)", # size of markers
                    animation_frame='Year')
fig.show()
```
